```python
import tensorflow as tf
import numpy as np
```

```python
class ContextualBandit:

    def __init__(self):
        self.active_bandit = 0  # state
        self.bandits = np.array([
             [0.2, 0.0, 0.1, -4.0],  # 4th arm best
             [0.1, -5.0, 1.0, 0.25],  # 2nd arm best
             [-3.5, 2.0, 3.2, 6.4]  # 1st arm best
        ])
        self.num_bandits, self.num_actions = self.bandits.shape

    def get_bandit(self):
        self.active_bandit = np.random.randint(0, self.num_bandits)
        return self.active_bandit

    def pull(self, arm):
        bandit = self.bandits[self.active_bandit, arm]
        return 1 if np.random.randn(1) > bandit else -1
```

```python
env = ContextualBandit()
num_actions = 4
contexts = 3
context = env.get_bandit()
context_one_hot = tf.expand_dims(tf.one_hot(context, contexts), 0)
W = tf.Variable(tf.random.normal([contexts, num_actions]))
```

```python
def predict(context):
    print(context_one_hot)
    output = tf.sigmoid(tf.matmul(context_one_hot, W))
    best_action = tf.argmax(output, axis=1)
    return best_action
```

```python
def random_or_predict(epsilon, context):
        if np.random.rand(1) < epsilon:
            return np.random.randint(num_actions)
        else:
            return predict(context)
```
